```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# dataset
df=pd.read_csv('products_asos.csv')
df.head()
```

```python
df['price']=pd.to_numeric(df['price'], errors='coerce')
df=df.dropna(subset=['price'])
df.head()
```

```python
df['description']=df['description'].astype(str) # forcing description to string
# getting last names from description column \
def get_brand(text):
  if 'by' in text:
    try:
      return text.split('by ')[1].split(' ')[0]
    except:
      return "unkown"
df['brand_row']=df['description'].apply(get_brand)

```

```python
df.head()
```

```python
brand_map={
    'New': 'New Look',
    'River': 'River Island',
    'Miss':'Miss Selfridge',
    'TopshopWelcome':'Topshop'
}
df['brand']=df['brand_row'].map(brand_map).fillna(df['brand_row'])
brand_counts=df['brand'].value_counts()
valid_brands=brand_counts[brand_counts>5].index
df_clean=df[df['brand'].isin(valid_brands)].copy()

print(df_clean['brand'].value_counts().head())
```

```python

# to check items which are out of stock
#function to analyze stockouts
def calculate_phantom_revenue(size_str):

    if not isinstance(size_str,str):
        return 0, 0.0
    # split 'UK 6, UK 8 - Out of Stock' into list
    sizes=size_str.split(',')
    total_sizes=len(sizes)

    # count how any items are out of stock
    out_of_stock_count= size_str.count('Out of stock')

    #calculate rate (0.0 - 1.0)
    rate= out_of_stock_count/ total_sizes if total_sizes>0 else 0.0

    return out_of_stock_count, rate

metrics= df_clean['size'].apply(lambda x: calculate_phantom_revenue(x))

df_clean['stockout_count']= [x[0] for x in metrics]
df_clean['stockout_rate']= [x[1] for x in metrics]

df_clean['lost_revenue']= df_clean['price'] * df_clean['stockout_count']

cols=['brand', 'name', 'price', 'stockout_count', 'lost_revenue']
print(df_clean.sort_values(by='lost_revenue', ascending=False).head(5)[cols])

```

```python
# Final recommedation
# Try to aggregate data together and giving stragetic advice on different brands

brand_stragetic=df_clean.groupby('brand').agg({
    'price': 'mean',
    'stockout_count': 'sum',
    'stockout_rate': 'mean',
    'lost_revenue': 'sum',
    'name': 'count'
}).reset_index()

brand_stragetic=brand_stragetic[brand_stragetic['name']>10]

plt.figure(figsize=(12,8))
sns.scatterplot(
    data=brand_stragetic,
    x='price',
    y='stockout_rate',
    size='lost_revenue',
    sizes=(50,500),
    alpha=0.7,
    palette='viridis'
)
winners= brand_stragetic[
    (brand_stragetic['price']>40) &
    (brand_stragetic['stockout_rate']>0.4)
]

for i in range(len(winners)):
  plt.text(
      winners.iloc[i]['price']+1,
      winners.iloc[i]['stockout_rate'],
      winners.iloc[i]['brand']
  )
plt.title('brand strategic analysis')
plt.xlabel('average price')
plt.ylabel('stockout rate')
plt.axvline(x=40, color='red', linestyle='--')
plt.axhline(y=0.4, color='red', linestyle='--')
plt.show()
```
